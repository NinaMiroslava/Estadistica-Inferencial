## **Que es una regresin lineal**

Es una tcnica de anlisis de datos que predice el valor de datos desconocidos mediante el uso de otro valor de datos relacionado y conocido. Modela maticamente la variable dependiente y la variable independiente como una ecuacin lineal. 


La regresin lineal se utiliza para producir el valor de una variable segn el valor de otra por lo que existen diferentes tipos de anlisis de regresin lineal.


1 Modelo de regresin lineal simple

2 Modelo de regresin lineal mltiple

3 Modelo de regresin no lineal


La importancia de la regresin lineal es que permite predecir el comportamiento de una variable, por lo que tiene presunciones como linealidad de la relacin la normalidad la aleatoriedad de la muestra y homogeneidad de las varianzas.

## **Regresin** **lineal**

Para este modelo estadistico la variable independiente sern los precios de una serie financiera de Microsoft, en donde la primera observacin fue realizada el 6 de enero del 2017 y la ultima el 31 de enero del 2022. En cuanto a la variable dependiente del modelo, ser tomado el volumen de trading que registra la serie 

```python
serieFin= pd.read_csv('Microsoft.csv')
xI=serieFin['Cierre'] #Variable independiente/explicativa de la regresin
yD=pd.to_numeric(serieFin['Vol.'], errors='coerce') #Variable dependiente/responsiva de la regresin
```

```python
serieFin.head(1276)
```

```python
ModeloOSL=smf.ols('y~x',serieFin) #le llamo "ModeloOSL", de la librearia de stats uso el modelo OLS para hacer la regresin.
Resultados=ModeloOSL.fit() #pido el modelo ajustado a la linea y lo llamo resultados
print(Resultados.summary()) #imprime el resumen de todos los resultados estadisticos
```

```python
print(Resultados.params) #Me da los parametros principales de la regresin
```

De acuerdo con la regresin obtenida calculada bajo el modelo de cuadrados ordinarios sabemos que el intercepto es 2.698986e+07 y que por un aumento unitario en el precio de la accin (variable independiente) el volumen (variable dependiente) aumentar en 1.170705e+04 unidades aproximadamente.


Siguiendo con el anlisis de los resultados, dado que el resultado de r2 es sumamente pequeo (0.005), podemos saber que el modelo no es estadisticamente significativo, pues la variable independiente explica en forma muy superficial la variable de la variable independiente.
```python
plt.title ('Volumen de acciones de Microsoft contra su precio de cierre')
plt.ylabel ('Volumen de acciones')
plt.xlabel ('Precio Microsoft')
plt.plot(Resultados.fittedvalues,color='red',linestyle='-')
plt.savefig('my_scatterplot.svg',dpi=100) #dpi es la resolucin con la que se guardara la grafica
```
