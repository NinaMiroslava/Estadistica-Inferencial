## **Que es una regresión lineal**

Es una técnica de análisis de datos que predice el valor de datos desconocidos mediante el uso de otro valor de datos relacionado y conocido. Modela maticamente la variable dependiente y la variable independiente como una ecuación lineal. 


La regresión lineal se utiliza para producir el valor de una variable según el valor de otra por lo que existen diferentes tipos de análisis de regresión lineal.


1 Modelo de regresión lineal simple

2 Modelo de regresión lineal múltiple

3 Modelo de regresión no lineal


La importancia de la regresión lineal es que permite predecir el comportamiento de una variable, por lo que tiene presunciones como linealidad de la relación la normalidad la aleatoriedad de la muestra y homogeneidad de las varianzas.

## **Regresión** **lineal**

Para este modelo estadistico la variable independiente serán los precios de una serie financiera de Microsoft, en donde la primera observación fue realizada el 6 de enero del 2017 y la ultima el 31 de enero del 2022. En cuanto a la variable dependiente del modelo, será tomado el volumen de trading que registra la serie 

```python
serieFin= pd.read_csv('Microsoft.csv')
xI=serieFin['Cierre'] #Variable independiente/explicativa de la regresión
yD=pd.to_numeric(serieFin['Vol.'], errors='coerce') #Variable dependiente/responsiva de la regresión
```

```python
serieFin.head(1276)
```

```python
ModeloOSL=smf.ols('y~x',serieFin) #le llamo "ModeloOSL", de la librearia de stats uso el modelo OLS para hacer la regresión.
Resultados=ModeloOSL.fit() #pido el modelo ajustado a la linea y lo llamo resultados
print(Resultados.summary()) #imprime el resumen de todos los resultados estadisticos
```

```python
print(Resultados.params) #Me da los parametros principales de la regresión
```

De acuerdo con la regresión obtenida calculada bajo el modelo de cuadrados ordinarios sabemos que el intercepto es 2.698986e+07 y que por un aumento unitario en el precio de la acción (variable independiente) el volumen (variable dependiente) aumentará en 1.170705e+04 unidades aproximadamente.


Siguiendo con el análisis de los resultados, dado que el resultado de r2 es sumamente pequeño (0.005), podemos saber que el modelo no es estadisticamente significativo, pues la variable independiente explica en forma muy superficial la variable de la variable independiente. Debido a lo anterior procederemos a una segunda regresión evaluada bajo el mismo modelo (OLS), sin embargo ahora será una regresión multiple añadiendo como x2 la varianza de la acción de acuerdo a la serie financiera de Microsoft recuperada de la bolsa de Nueva York (NYSE)

```python
plt.title ('Volumen de acciones de Microsoft contra su precio de cierre')
plt.ylabel ('Volumen de acciones')
plt.xlabel ('Precio Microsoft')
plt.plot(Resultados.fittedvalues,color='red',linestyle='-')
plt.savefig('my_scatterplot.svg',dpi=100) #dpi es la resolución con la que se guardara la grafica
```
